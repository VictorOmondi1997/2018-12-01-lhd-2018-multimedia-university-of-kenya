# 2018-12-01 Local Hack Day 2018 Multimedia University of kenya

```python
import pandas as pd
from csv import reader

# open the csv file
opened_file = open('2018-12-01-lhd-2018-multimedia-university-of-kenya.csv')
read_file = reader(opened_file)

# enter the data into an lhd_data list
lhd_data = list(read_file)
```

## Column Name

```python
print(lhd_data[0])
```

```python
print('Index \t Column Name \n--------------------\n')
for column_name in lhd_data[0]:
    print(lhd_data[0].index(column_name), ' \t', column_name.capitalize())
```

## Names of Attendees and their contact details

```python
for data in lhd_data[1:]:
    name = data[4] + " " + data [5]
    email = data[3]
    phone_number = data[14]
    
    print(name, '\t\t', email, '\t\t\t', phone_number)
```

```python
def frequency_table(column_index):
    frequency = {}
    for data in lhd_data[1:]:
        value_data = data[column_index]
        if value_data in frequency:
            frequency[value_data] += 1
        else:
            frequency[value_data] = 1
    return frequency
```
